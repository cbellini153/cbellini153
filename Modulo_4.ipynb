```python
# 1. Eliminar columnas irrelevantes
df = df.drop([
    'Attraction Index',
    'Normalised Attraction Index',
    'Restraunt Index',
    'Normalised Restraunt Index'
    ], axis=1)

#df = df.drop(columns = {'Attraction Index',
#    'Normalised Attraction Index',
#    'Restraunt Index',
#    'Normalised Restraunt Index'})

#df.drop(columns = {'Attraction Index',
#    'Normalised Attraction Index',
#    'Restraunt Index',
#    'Normalised Restraunt Index'}, inplace = True)
```

```python
# 2. Columna Bedrooms_per_person
df['Bedrooms_per_person'] = df['Bedrooms']/df['Person Capacity']
```

```python
# 3. Elimino precios mayores a 10.000
df = df[df.Price < 10000]
```

```python
# 1. Scatter plot enre columnas de distancias
df.plot.scatter(x = 'City Center (km)', y = 'Metro Distance (km)', s = 100);
```

```python
# 2. Análisis por ciudad
df.groupby('City').agg({'Guest Satisfaction':['count', 'mean']})
```

```python
df_cities = df.groupby('City').agg({'Guest Satisfaction':['count', 'mean']})
df_cities = df_cities.reset_index()
df = df.merge(df_cities, on='City')
```

```python
# 1. Cálculo del benchmark

benchmark = df['Guest Satisfaction'].mean() + df['Guest Satisfaction'].std()/2
benchmark
```

```python
# 2. Armado de una regla

mask = (df.City.isin(['Athens', 'Budapest'])) & (df['Cleanliness Rating'] == 10) & (df['Person Capacity'] >= 4)

print(len(df[mask]), df[mask]['Guest Satisfaction'].mean())
```

```python
df_regla = df[mask]

print(df_regla['Price'].mean(), df['Price'].mean())
```
